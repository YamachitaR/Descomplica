```python
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.optimize import minimize
```

```python

# 1. Simulação de Dados (Preço x Quantidade Vendida)
np.random.seed(42)
precos = np.linspace(10, 100, 20)  # Preços simulados
quantidades = 120 - 0.9 * precos + np.random.normal(0, 5, len(precos))  # Qd = a - bP + erro


# Visualizando os dados simulados
plt.scatter(precos, quantidades, color='blue', label='Dados Simulados')
plt.xlabel('Preço')
plt.ylabel('Quantidade Demandada')
plt.title('Dados Simulados: Preço vs Quantidade Demandada')
plt.legend()
plt.show()
```

```python
# 2. Regressão Linear para Estimar a Função Demanda
precos_reshape = precos.reshape(-1, 1)
modelo = LinearRegression()
modelo.fit(precos_reshape, quantidades)

a = modelo.intercept_  # Intercepto (a)
b = modelo.coef_[0]  # Coeficiente de inclinação (b)

print(f"Intercepto (a): {a:.2f}")
print(f"Inclinação (b): {b:.2f}")
```

```python
# Função de Demanda Estimada: Qd = a - bP
def demanda(preco):
    return a + b * preco

# 3. Função de Lucro (considerando custo zero para simplificação)
def receita(preco):
    return preco * demanda(preco)

def custo(preco):
    return 2000 + 2 * demanda(preco)


def lucro(preco):
    return receita(preco) -  custo(preco)
```

```python
# 4. Encontrando o Preço Ótimo para Maximizar o Lucro
resultado = minimize(lambda p: -lucro(p), x0=[50], bounds=[(0, 200)])
preco_otimo = resultado.x[0]

# 5. Visualizando a Função Lucro e o Preço Ótimo
precos_range = np.linspace(10, 120, 100)
lucros = [lucro(p) for p in precos_range]

plt.plot(precos_range, lucros, label='Função Lucro', color='green')
plt.axvline(x=preco_otimo, color='red', linestyle='--', label=f'Preço Ótimo = R${preco_otimo:.2f}')
plt.xlabel('Preço')
plt.ylabel('Lucro')
plt.title('Maximização do Lucro')
plt.legend()
plt.show()
```
